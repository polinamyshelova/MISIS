```python
import telebot, urllib3
bot = telebot.TeleBot('TOKEN')
http = urllib3.PoolManager()

@bot.message_handler(commands=['start'])
def start(message):
    name = f'{message.from_user.first_name} {message.from_user.last_name}'
    bot.send_message(message.chat.id, f'<b>Хэй, {name}</b>! Напиши ссылку сайта, чтобы узнать его доступность.', parse_mode='html')

@bot.message_handler(content_types=['text'])
def get_message(message):
    URL = message.text
    request = http.request('GET', URL)
    if request.status == 200:
        bot.send_message(message.chat.id, f'Йоу! Сайт {URL} доступен!')
    else:
        bot.send_message(message.chat.id, f'Сайт {URL} недоступен...')

bot.polling(none_stop=True, interval=0)
```
